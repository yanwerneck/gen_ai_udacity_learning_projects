```python
import os
from pydantic import BaseModel, Field
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate

class RealStateListing(BaseModel):
    neighborhood: str = Field(description="Neighborhood of the listed house")
    price: float = Field(description="Sale price of the house")
    bedrooms: int = Field(description="The number of bedrooms in the house")
    bathrooms: int = Field(description="The number of bathrooms in the house")
    size: float = Field(description="The size of  the house in square feet")
    house_description: str = Field(description="A description of the given house")
    neighborhood_description: str = Field(description="A description of the given house's neighborhood")

class AllListings(BaseModel):
    houses: list[RealStateListing]

    
EXAMPLE = """
Neighborhood: Green Oaks
Price: $800,000
Bedrooms: 3
Bathrooms: 2
House Size: 2,000 sqft

Description: Welcome to this eco-friendly oasis nestled in the heart of Green Oaks. This charming 3-bedroom, 2-bathroom home boasts energy-efficient features such as solar panels and a well-insulated structure. Natural light floods the living spaces, highlighting the beautiful hardwood floors and eco-conscious finishes. The open-concept kitchen and dining area lead to a spacious backyard with a vegetable garden, perfect for the eco-conscious family. Embrace sustainable living without compromising on style in this Green Oaks gem.

Neighborhood Description: Green Oaks is a close-knit, environmentally-conscious community with access to organic grocery stores, community gardens, and bike paths. Take a stroll through the nearby Green Oaks Park or grab a cup of coffee at the cozy Green Bean Cafe. With easy access to public transportation and bike lanes, commuting is a breeze.
"""


NUMBER_OF_LISTINGS = 100

MODEL_NAME = "gpt-3.5-turbo"
```

```python
llm = ChatOpenAI(model_name=MODEL_NAME, temperature=1, max_tokens=2000)
structured_llm = llm.with_structured_output(AllListings)
```

```python
prompt = PromptTemplate(
    template="{query}\nContext: {example}",
    input_variables=["query", "example"],
)

fake_listings_query = f"""You are a seasoned realtor and you are curating a real state listing. 
Generate real estate {NUMBER_OF_LISTINGS} listings. 
The fields are:
neighborhood: str = Field(description="Neighborhood of the listed house")
price: float = Field(description="Sale price of the house")
bedrooms: int = Field(description="The number of bedrooms in the house")
bathrooms: int = Field(description="The number of bathrooms in the house")
size: float = Field(description="The size of  the house in square feet")
house_description: str = Field(description="A description of the given house")
neighborhood_description: str = Field(description="A description of the given house's neighborhood")
Produce descriptions of various properties very similar to reality.
Make sure to repeat some of the 
Format the output in a json format
"""

real_state_query = prompt.format(query = fake_listings_query, example = EXAMPLE)
```

```python
listings = llm.invoke(fake_listings_query) 
```

```python
listings.to_json()
```
